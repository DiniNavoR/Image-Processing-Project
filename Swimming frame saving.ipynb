```python
import cv2
import os

# Path to the video
video_path = r"C:\Users\dinit\OneDrive\Documents\University\Second_year_First_semester\COM 2304 Computer Graphic and Image Processing\Assignment\video 1.mp4"

# Folder to save frames
output_folder = r"C:\Users\dinit\OneDrive\Documents\University\Second_year_First_semester\COM 2304 Computer Graphic and Image Processing\Assignment\frames1"

# Create the folder if it does not exist
os.makedirs(output_folder, exist_ok=True)

# Open the video
video = cv2.VideoCapture(video_path)

frame_number = 1

while True:
    ret, frame = video.read()

    if not ret:
        break

    # Convert frame to grayscale
    gray_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    # Save grayscale frame
    filename = os.path.join(output_folder, f"frame_{frame_number}.jpg")
    cv2.imwrite(filename, gray_frame)

    frame_number += 1

video.release()

print("Grayscale frames of first video saved successfully!")
```
