```python
import cv2
import numpy as np
import os

# Folder containing enhanced frames
input_folder = r"C:\Users\dinit\OneDrive\Documents\University\Second_year_First_semester\COM 2304 Computer Graphic and Image Processing\Assignment\enhanced_frames1"

# Folder to save segmented images
output_folder = r"C:\Users\dinit\OneDrive\Documents\University\Second_year_First_semester\COM 2304 Computer Graphic and Image Processing\Assignment\segmented_frames1"

# Create output folder
os.makedirs(output_folder, exist_ok=True)

# Loop through all images
for filename in os.listdir(input_folder):

    if filename.endswith(".jpg"):

        image_path = os.path.join(input_folder, filename)

        # Read image
        image = cv2.imread(image_path)

        # Convert to HSV color space
        hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

        # Water color range (blue/green)
        lower_water = np.array([80, 50, 50])
        upper_water = np.array([140, 255, 255])

        # Create mask for water
        mask = cv2.inRange(hsv, lower_water, upper_water)

        # Invert mask to highlight swimmer
        swimmer_mask = cv2.bitwise_not(mask)

        # Morphological operations to remove noise
        kernel = np.ones((5,5), np.uint8)
        cleaned = cv2.morphologyEx(swimmer_mask, cv2.MORPH_OPEN, kernel)
        cleaned = cv2.morphologyEx(cleaned, cv2.MORPH_CLOSE, kernel)

        # Apply mask to original image
        segmented = cv2.bitwise_and(image, image, mask=cleaned)

        # Save segmented image
        save_path = os.path.join(output_folder, filename)
        cv2.imwrite(save_path, segmented)

print("Segmentation completed for all frames of video 1!")
```
