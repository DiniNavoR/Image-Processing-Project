```python
import cv2
import os

# Folder containing noisy images
input_folder = r"C:\Users\dinit\OneDrive\Documents\University\Second_year_First_semester\COM 2304 Computer Graphic and Image Processing\Assignment\frames2"

# Folder to save denoised images
output_folder = r"C:\Users\dinit\OneDrive\Documents\University\Second_year_First_semester\COM 2304 Computer Graphic and Image Processing\Assignment\enhanced_frames2"

# Create output folder if it doesn't exist
os.makedirs(output_folder, exist_ok=True)

import numpy as np

kernel = np.array([[0,-1,0],
                   [-1,5,-1],
                   [0,-1,0]])

# Loop through all images in the folder
for filename in os.listdir(input_folder):

    if filename.endswith(".jpg"):

        image_path = os.path.join(input_folder, filename)

        # Read the image
        image = cv2.imread(image_path)

        # Apply Gaussian Blur (removes Gaussian noise)
        denoised = cv2.GaussianBlur(image, (5,5), 0)

        # Apply Gaussian blur
        denoised = cv2.GaussianBlur(image, (5,5), 0)
        
        # Convert to grayscale
        gray = cv2.cvtColor(denoised, cv2.COLOR_BGR2GRAY)
        
        # Apply histogram equalization
        histogram = cv2.equalizeHist(gray)

        #sharpened image
        sharpened = cv2.filter2D(histogram, -1, kernel)

        # Save the processed image
        save_path = os.path.join(output_folder, filename)
        cv2.imwrite(save_path, sharpened)

print("Enhancing completed!")
```
