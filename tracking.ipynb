```python
import cv2
import numpy as np

# 1. Initialize Video
cap = cv2.VideoCapture('video 1.mp4')
width  = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps    = cap.get(cv2.CAP_PROP_FPS)

# Use XVID/AVI for the best results in OpenCV
out = cv2.VideoWriter('swimmer_tracked_box.avi', cv2.VideoWriter_fourcc(*'XVID'), fps, (width, height))

# 2. Background Subtractor (Higher threshold = steadier box)
backSub = cv2.createBackgroundSubtractorMOG2(history=500, varThreshold=100, detectShadows=False)

print("Processing: Tracking the main swimmer...")

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    # 3. Focus on the swimming lanes (Cropping the noise from crowd/deck)
    # We create a Region of Interest (ROI) mask
    mask_roi = np.zeros(frame.shape[:2], dtype="uint8")
    cv2.rectangle(mask_roi, (0, int(height*0.2)), (width, int(height*0.8)), 255, -1)
    
    # Apply Subtraction
    fg_mask = backSub.apply(frame)
    fg_mask = cv2.bitwise_and(fg_mask, fg_mask, mask=mask_roi)

    # 4. Morphological Operations (Closing the gaps in the 'blob')
    # This helps turn the swimmer's head, arms, and body into ONE single box
    kernel = np.ones((15, 15), np.uint8) 
    fg_mask = cv2.morphologyEx(fg_mask, cv2.MORPH_CLOSE, kernel)
    fg_mask = cv2.dilate(fg_mask, kernel, iterations=1)

    # 5. Find Contours
    contours, _ = cv2.findContours(fg_mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    
    if contours:
        # Pick the largest moving object (the main swimmer)
        c = max(contours, key=cv2.contourArea)
        
        if cv2.contourArea(c) > 1500: # Adjust size threshold if needed
            x, y, w, h = cv2.boundingRect(c)
            
            # 6. Draw the Box
            # Color: Green (0, 255, 0), Thickness: 3
            cv2.rectangle(frame, (x, y), (x + w, y + h), (0, 255, 0), 3)
            
            # Optional: Add a label that follows the box
            cv2.putText(frame, "TRACKING", (x, y - 10), 
                        cv2.FONT_HERSHEY_SIMPLEX, 0.6, (0, 255, 0), 2)

    out.write(frame)

# 7. Release and Save
cap.release()
out.release()
print("Done! You can now download 'swimmer video1.avi'")
```
